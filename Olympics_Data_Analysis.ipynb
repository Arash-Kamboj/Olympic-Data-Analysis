**8.Data Type Conversion:**

We ensure that relevant columns are of the correct data types for analysis.


---



```python
# Data Type Conversion

df['Avg_Temp'] = pd.to_numeric(df['Avg_Temp'], errors='coerce')  # Convert to numeric, converting errors to NaN
print(f"Data type of 'Avg_Temp' column : {df['Avg_Temp'].dtypes}")
```

Data sanitisation after the data conversion of 'Avg_Temp'. Check again if any new null values are created.


---



```python
df.isnull().sum()
```

**9.Handling missing values.**

Impute missing Values in 'Avg_Temp' using median.

---



```python
 # Replace missing values in 'Avg_Temp' with median
df['Avg_Temp'] = df['Avg_Temp'].fillna(df['Avg_Temp'].median())
df.isnull().sum()
```

**10. Imputing missing values.**

Impute lissing values in 'Height and 'Weight' based on the corresponding 'Age' value
.

---



```python
# Creating avg_height variable
avg_height= df.groupby('Age')['Height'].median().to_dict()

# Iterate rows to replace missing values in 'Height'
for index, row in df.iterrows():
    if pd.isnull(row['Height']):
        avg_age = row['Age']
        if avg_age in avg_height:
            df.at[index, 'Height'] = avg_height[avg_age]


# Creating avg_weight variable
avg_weight= df.groupby('Age')['Weight'].median().to_dict()

# Iterate rows to replace missing values in 'Weight'
for index, row in df.iterrows():
    if pd.isnull(row['Weight']):
        avg_age = row['Age']
        if avg_age in avg_weight:
           df.at[index, 'Weight'] = avg_weight[avg_age]

df.isnull().sum()
```

```python
# Show cleaned data summary
data_type = pd.DataFrame(df.dtypes, columns=['Data Type'])
missing = pd.DataFrame(df.isnull().sum(), columns=['Missing Values'])
missing_percentage = pd.DataFrame(round(df.isnull().mean() * 100,2), columns=['Missing Percentage'])
unique = pd.DataFrame(df.nunique(), columns=['Unique Values'])
sample = pd.DataFrame(df.head().T)

summary = pd.concat([data_type, missing, missing_percentage, unique, sample], axis=1)
summary.rename(columns={0: 'Row_1',
                        1: 'Row_2',
                        2: 'Row_3',
                        3: 'Row_4',
                        4: 'Row_5'
                       }, inplace=True)
summary
```

**Finding Team with highest number of medals.**


---



```python
medal_counts = df.groupby('Team')['Medal'].sum()

# Finding the country with the maximum sum of medals
country_with_max_medals = medal_counts.idxmax()
max_medals = medal_counts.max()

print("Country with maximum medals:", country_with_max_medals)
print("Number of medals won:", max_medals)
```

**Finding Sport with maximum number of medals.**


---



```python
medal_counts = df.groupby('Sport')['Medal'].sum()

# Finding the country with the maximum sum of medals
Sport_with_max_medals = medal_counts.idxmax()
max_medals = medal_counts.max()

print("Sport with maximum medals:", Sport_with_max_medals)
print("Number of medals won:", max_medals)
```

Finding Top 5 countries and their visual representation.

```python
sorted_medal_counts = medal_counts.sort_values(ascending=False)

# Taking the top 5 teams
top_5_teams = sorted_medal_counts.head(5)
print("Top 5 teams based on the number of medals won:")
print(top_5_teams)

# Create a bar plot to visualize the distribution
plt.figure(figsize=(10,5))
sns.set_style("darkgrid")
sns.barplot(top_5_teams, palette='mako')
plt.title('Distribution of Origin', fontsize=14, fontweight='bold')
plt.xlabel('Origin of Owner', fontweight='bold')
plt.ylabel('Count', fontweight='bold')
plt.show()

```

**Analysing relation between Weight and Age on gender basis.**

**Correlation Analysis**

 Correlation analysis on all numerical variables and its inferences.

```python
df = pd.read_csv("Olympics.csv")
round(df.corr(),2)
```

**Visual representation of Correlational analysis.**

Each cell in the heatmap represents the correlation coefficient between two variables. The color intensity indicates the strength and direction of the correlation: darker colors indicate stronger correlations, while lighter colors indicate weaker correlations. The annotation within each cell displays the correlation coefficient value.

```python
plt.figure(figsize=(8,4))
sns.heatmap(round(df.corr(),2), annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})
plt.title('Correlation Heatmap')
plt.show()

print("Most of the variables are not significantly correlated\nbut we could see that height and weight are highly correlated with correlation coefficient ( r) as 0.80.\nThis is logical as we know that biologically also if height increases weight increases.")
```

**Categorical Analysis**

In this code I tried to found out top 3 participating nations in 'Swimming' and then I created a separate dataset based on this.


---



```python
df_swim = df[df['Sport']=='Swimming']
df_swim['Team'].value_counts()
```

```python
df_swim_top3 = df_swim.loc[df_swim['Team'].isin(['United States','Germany','Australia'])]
df_swim_top3
```

```python
plt.figure(figsize=(6,4))
df_swim_top3.groupby('Team').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
# Create a boxplot on the first subplot

sns.boxplot(x="Team", y="Height", data=df_swim_top3, ax=axes[0])
axes[0].set_xlabel("Team")
axes[0].set_ylabel("Height")
axes[0].set_title("Boxplot")

# Create a histogram on the second subplot

sns.histplot(data=df_swim_top3, x="Height", bins=20, color='Red', kde=True, ax=axes[1])
axes[1].set_xlabel("Height")
axes[1].set_ylabel("Frequency")
axes[1].set_title("Histogram")

plt.tight_layout()
plt.show()

```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
# Create a boxplot on the first subplot

sns.boxplot(x="Team", y="Weight", data=df_swim_top3, ax=axes[0])
axes[0].set_xlabel("Team")
axes[0].set_ylabel("Weight")
axes[0].set_title("Boxplot")

# Create a histogram on the second subplot

sns.histplot(data=df_swim_top3, x="Weight", bins=20, color='Green', kde=True, ax=axes[1])
axes[1].set_xlabel("Weight")
axes[1].set_ylabel("Frequency")
axes[1].set_title("Histogram")

plt.tight_layout()
plt.show()
```

**Explanation**

1. Overall height looks slightly left skewed whereas the United States and Germany almost have similar distribution whereas distribution for Australia is slightly lower.

2. Overall weight looks slightly left skewed whereas Australia and Germany almost have similar distribution whereas distribution for United States is slightly higher.
